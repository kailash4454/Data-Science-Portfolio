# Exercise 1: Prime Number Checker

```python
# Taking input from user
num = int(input("Enter a number: "))

# Prime numbers are greater than 1
if num <= 1:
    print(num, "is not a Prime Number")

else:
    is_prime = True

    # Check divisibility from 2 to square root of number
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

    # Final result
    if is_prime:
        print(num, "is a Prime Number")
    else:
        print(num, "is not a Prime Number")
```

# Exercise 2: Product of Random Numbers

```python
# Import Random Module
import random

# Generate two random numbers between 1 and 20
num1 = random.randint(1, 20)
num2 = random.randint(1, 20)

# Actual product
correct_answer = num1 * num2

# Ask user to calculate the product
user_answer = int(input(f"What is {num1} × {num2}? "))

# Check user's answer
if user_answer == correct_answer:
    print("Correct! Well done.")
else:
    print(f"Wrong! The correct answer is {correct_answer}.")
```

# Exercise 4: Hospital Billing System

```python
# Fixed hospital charges
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_cost = 300

# Patient details
days_stayed = 4
number_of_tests = 3
medicine_charges = 2400

# Calculate room charges
room_charges = days_stayed * room_charge_per_day

# Calculate lab charges
lab_charges = number_of_tests * lab_test_cost

# Total bill before discount
total_bill = room_charges + doctor_fee + lab_charges + medicine_charges

# Apply discount if total bill exceeds ₹10,000
if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

# Final payable bill
final_bill = total_bill - discount

# Display results
print("Room Charges: ₹", room_charges)
print("Doctor Consultation Fee: ₹", doctor_fee)
print("Lab Test Charges: ₹", lab_charges)
print("Medicine Charges: ₹", medicine_charges)
print("Total Bill Before Discount: ₹", total_bill)
print("Discount Applied: ₹", discount)
print(f"Final Bill Amount: ₹{final_bill}")
```
