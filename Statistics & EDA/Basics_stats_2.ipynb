# Step 1 — Load Patient Dataset + Summary using info()

```python
import pandas as pd
import numpy as np

# Load patient dataset
patient_df = pd.read_csv("Patient_Data.csv")
```

```python
# Display first 5 rows
print("First 5 Rows of Patient Dataset:\n")
print(patient_df.head())
```

```python
# Shape of dataset
print("\nShape of Dataset:")
print(patient_df.shape)
```

```python
# Column names
print("\nColumn Names:")
print(patient_df.columns)
```

```python
# Dataset summary using info()
print("\nDataset Info:\n")
print(patient_df.info())
```

# Step 2 — Select Only Billing-Related Columns

```python
billing_related_columns = [
    "PatientID",
    "Department",
    "Doctor",
    "BillAmount"
]

billing_df = patient_df[billing_related_columns]

print("Billing Related Columns Dataset:\n")
print(billing_df.head())
```

```python
print("Shape After Selecting Billing Columns:")
print(billing_df.shape)
```

# Step 3 — Drop Administrative Columns

```python
columns_to_drop = [
    "ReceptionistID",
    "CheckInTime"
]

clean_patient_df = patient_df.drop(
    columns=columns_to_drop,
    errors="ignore"
)

print("Dataset After Dropping Administrative Columns:\n")
print(clean_patient_df.head())
```

```python
print("Shape After Dropping Columns:")
print(clean_patient_df.shape)
```

# Step 4 — GroupBy → Total Bill Amount Per Department

```python
department_bill_summary = patient_df.groupby(
    "Department",
    as_index=False
)["BillAmount"].sum()

print("Total Bill Amount Per Department:\n")
print(department_bill_summary)
```

```python
# Sort Highest Revenue First

department_bill_summary = department_bill_summary.sort_values(
    by="BillAmount",
    ascending=False
)

print(department_bill_summary)
```

# Step 5 — Remove Duplicate Patient Records Based on PatientID

```python
patient_df_no_duplicates = patient_df.drop_duplicates(
    subset="PatientID",
    keep="first"
)

print("Dataset After Removing Duplicate Patient Records:\n")
print(patient_df_no_duplicates.head())
```

```python
print("Rows Before Removing Duplicates:", patient_df.shape[0])
print("Rows After Removing Duplicates:", patient_df_no_duplicates.shape[0])
```

# Step 6 — Fill Missing BillAmount Values Using Mean

```python
print("Missing Values in BillAmount Before Filling:")
print(patient_df["BillAmount"].isnull().sum())
```

```python
patient_df["BillAmount"] = patient_df["BillAmount"].fillna(
    patient_df["BillAmount"].mean()
)

print("Dataset After Filling Missing BillAmount:\n")
print(patient_df.head())
```

```python
print("Missing Values in BillAmount After Filling:")
print(patient_df["BillAmount"].isnull().sum())
```

# Step 7 — Merge Billing Dataset with Patient Dataset on PatientID

```python
# Load billing dataset

billing_data = pd.read_csv("Billing_Data.csv")

print("First 5 Rows of Billing Dataset:\n")
print(billing_data.head())
```

```python
# Merge Patient and Billing Dataset

merged_df = pd.merge(
    patient_df,
    billing_data,
    on="PatientID",
    how="inner",
    suffixes=("_Patient", "_Billing")
)

print("Merged Dataset:\n")
print(merged_df.head())
```

```python
print("Patient Dataset Shape:", patient_df.shape)
print("Billing Dataset Shape:", billing_data.shape)
print("Merged Dataset Shape:", merged_df.shape)
```

# Step 8 — Concatenate New Patients for Current Week (Row-wise)

```python
# New Patients for Current Week

new_patients = pd.DataFrame({
    "PatientID": [201, 202],
    "Name": ["Kailash", "Sonika"],
    "Department": ["Orthopedic", "Pediatrics"],
    "Doctor": ["Dr. Mehta", "Dr. Khan"],
    "BillAmount": [8000, 6500],
    "ReceptionistID": [1, 3],
    "CheckInTime": ["2023-01-13 12:00", "2023-01-10 09:00"]
})

print("New Weekly Patients:\n")
print(new_patients)
```

```python
# Row-wise Concatenation

updated_patient_df = pd.concat(
    [patient_df, new_patients],
    axis=0,
    ignore_index=True,
    sort=False
)

print("Dataset After Adding New Patients:\n")
print(updated_patient_df.tail())
```

```python
print("Original Dataset Shape:", patient_df.shape)
print("Updated Dataset Shape:", updated_patient_df.shape)
```

# Step 9 — Concatenate New Billing Columns (Column-wise)

```python
additional_billing_info = pd.DataFrame({
    "InsuranceCovered": ["Yes", "No", "Yes", "No", "Yes"],
    "FinalAmount": [4500, 7000, 5200, 6100, 3900]
})

print("Additional Billing Columns:\n")
print(additional_billing_info)
```

```python
print(patient_df.shape[0])
print(additional_billing_info.shape[0])
```

```python
# Column-wise Concatenation

final_patient_df = pd.concat(
    [patient_df.head(5), additional_billing_info],
    axis=1
)

print("Final Dataset After Adding New Billing Columns:\n")
print(final_patient_df)
```
