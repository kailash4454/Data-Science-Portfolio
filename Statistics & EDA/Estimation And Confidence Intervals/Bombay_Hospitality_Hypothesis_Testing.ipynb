# Step 1 — Understand the Business Problem + State Hypotheses

## Understand the Business Problem

The company provides the operating cost model:

W=1000 + 5X

Where:

W = Weekly Operating Cost

X = Number of units produced per week

The company assumes this model is correct.

## Given:

Average production:

μx = 600 units

Sample size:

n=25 restaurants

Observed sample mean weekly cost:

ˉx = 3050

## Restaurant owners claim:

Actual operating costs are higher

than the old model predicts.

## State the Hypotheses

Since owners claim:

Costs are higher

this becomes:

Right-Tailed Test

```python
# Hypothesis Statement

# Null Hypothesis (H0): μ ≤ μ0​
# The weekly operating cost has not increased.
# The original cost model is still accurate.

# Alternative Hypothesis (H1): μ > μ0​
# The weekly operating cost is higher than the model predicts.
# Restaurant owners' claim is correct.

# This is a Right-Tailed Z-Test
```

# Step 2 — Calculate Theoretical Mean Weekly Cost

```python
# Calculated Theoretical Mean Weekly Cost

X = 600  # Average units produced
mu_0 = 1000 + 5 * X

print("Theoretical Mean Weekly Cost (μ0):", mu_0)
```

## Important Observation

Observed sample mean:

xˉ = 3050

while theoretical mean becomes:

μ0 = 4000

**Sample mean is actually lower than theoretical mean**

not higher.

Restaurant owners claim:

Costs are higher

but sample mean shows:

Costs are lower

**This means our final hypothesis test may reject their claim.**

# Step 3 — Calculate Test Statistic (Z-Test)

```python
# Step 3: Calculate Test Statistic (Z-Test)

x_bar = 3050      # Sample mean
mu = 4000         # Theoretical mean
sigma = 125       # Population standard deviation
n = 25            # Sample size

z = (x_bar - mu) / (sigma / (n ** 0.5))

print("Z-Test Statistic:", z)
```

## Important Observation

This is a:

### **Very large negative Z-value**

This means:

**Sample mean is far below theoretical mean**

not above it.

This strongly goes against the restaurant owners’ claim.

If the claim was:

**Costs are higher**

we would expect:

**positive Z-value**

But we got:

**strongly negative Z-value**

This means:

the actual evidence says the opposite.

This is exactly why hypothesis testing protects businesses from wrong assumptions.


# Step 4 — Determine Probability + Compare with Alpha

```python
# Compare with Critical Value

alpha = 0.05
z_critical = 1.645   # Right-tailed test critical value
z = -38

print("Test Statistic (Z):", z)
print("Critical Value:", z_critical)

if z > z_critical:
    print("Reject Null Hypothesis (H0)")
else:
    print("Fail to Reject Null Hypothesis (H0)")
```

## Understanding Probability (P-Value Thinking)

Since:

z = −38

is extremely far on the left side,

for a:

**right-tailed test**

the p-value becomes:

**extremely large (close to 1)**

because we are testing for:

**increase**

but observed evidence shows:

**strong decrease**

So:

**p > 0.05**

This again means:

### Fail to reject H₀

# Step 5 — Final Decision (Reject / Fail to Reject H₀)

```python
# Since the calculated Z-value (-38)
# is less than the critical value (1.645),

# we Fail to Reject the Null Hypothesis (H0)

# There is not enough statistical evidence
# to support the claim that weekly operating costs
# are higher than the theoretical model predicts.
```

## Important Business Insight

**Fail to Reject H₀**

does NOT mean:

**H₀ is proven true**

It means:

**We do not have enough evidence against it**
