## Question 1: List Operations

```python
# Creating a list of 5 integers
numbers = [10, 20, 30, 40, 50]

print("Original List:", numbers)

# 1. append() → Adds one element at the end
numbers.append(60)
print("After append(60):", numbers)

# 2. extend() → Adds multiple elements
numbers.extend([70, 80])
print("After extend([70, 80]):", numbers)

# 3. insert() → Inserts element at specific position
numbers.insert(2, 25)
print("After insert(2, 25):", numbers)

# 4. remove() → Removes first matching value
numbers.remove(40)
print("After remove(40):", numbers)

# 5. pop() → Removes element by index (default last)
removed_item = numbers.pop()
print("After pop():", numbers)
print("Popped Item:", removed_item)

# 6. index() → Finds position of element
position = numbers.index(30)
print("Index of 30:", position)

# 7. count() → Counts occurrence of element
count_20 = numbers.count(20)
print("Count of 20:", count_20)

# 8. sort() → Sorts list in ascending order
numbers.sort()
print("After sort():", numbers)

# 9. reverse() → Reverses the list
numbers.reverse()
print("After reverse():", numbers)

# 10. clear() → Removes all elements
numbers.clear()
print("After clear():", numbers)
```

## Question 2: Tuple and Immutability


```python
# Creating a tuple with 3 student names
students = ("Kailash", "Riya", "Sonika")

print("Original Tuple:", students)
```

```python
# Trying to change the second name
students[1] = "Anjali"
```

The above code gives a **TypeError** because tuples are **immutable** in Python.

Once a tuple is created, its elements cannot be changed, added, or removed.

```python
print("Updated Tuple:", students)
```

## Question 3: Set with Duplicate Values

```python
# Creating a set with duplicate values
numbers = {10, 20, 30, 20, 40, 10, 50}

# Printing the set
print("Set Output:", numbers)
```

The duplicate values (10 and 20) are automatically removed because **sets** only store **unique elements**.

Also, the order of elements may be different because **sets** are **unordered** collections.

That is why the output contains only unique values and the sequence may change.

## Question 5: Voting Eligibility Program

```python
age = int(input("Enter your age: "))

if age < 0:
    print("Invalid age entered.")
elif age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

## Question 6: Marks to Grade Program

```python
# User Input marks
marks = int(input("Enter marks: "))

# Grade calculation
if marks < 0 or marks > 100:
    print("Invalid marks entered.")
elif marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 50:
    print("Grade: C")
else:
    print("Grade: Fail")
```

## Question 7: Positive, Even, Zero, Negative Check

```python
# Taking user input number as a variable
num = int(input("Enter a number: "))

# First check if number is positive
if num > 0:
    print("The number is positive.")

    # Nested condition to check even or odd
    if num % 2 == 0:
        print("It is also an even number.")
    else:
        print("It is an odd number.")

# If not positive, check for zero or negative
elif num == 0:
    print("The number is zero.")

else:
    print("The number is negative.")
```

## Question 8: Scalar, 1D Array, and 2D Array

```python
import numpy as np

# 1. Scalar (0-D Array)
scalar = np.array(5)

# 2. 1D Array with values 1 to 5
array_1d = np.array([1, 2, 3, 4, 5])

# 3. 2D Array (2x3) with values from 10 to 60
array_2d = np.arange(10, 70, 10).reshape(2, 3)

print("Scalar:\n", scalar)
print("\n1D Array:\n", array_1d)
print("\n2D Array:\n", array_2d)
```

## Question 10: NumPy Array to Pandas DataFrame

```python
import numpy as np
import pandas as pd

# Creating a 3x3 NumPy array
array_2d = np.arange(10, 100, 10).reshape(3, 3)

print("NumPy Array:")
print(array_2d)

# Converting into Pandas DataFrame
df = pd.DataFrame(array_2d, columns=["A", "B", "C"])

print("\nPandas DataFrame:")
print(df)
```

## Question 11: Full Pandas DataFrame Operations

```python
import pandas as pd
import numpy as np

# Creating DataFrame manually
data = {
    "Name": ["Rahul", "Priya", "Aman", "Sneha", "Karan",
             "Riya", "Vikas", "Neha", "Aman", "Sneha"],

    "Age": [22, 24, 23, 25, 26,
            21, 27, 24, 23, 25],

    "City": ["Delhi", "Mumbai", "Pune", "Delhi", "Chennai",
             "Kolkata", "Mumbai", "Pune", "Pune", "Delhi"],

    "Salary": [50000, 60000, np.nan, 55000, 70000,
               48000, 65000, 62000, np.nan, 55000]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n")
print(df)
```

```python
print("DataFrame Info:\n")
print(df.info())
```

```python
print("DataFrame Description:\n")
print(df.describe())
```

```python
print("Selected Columns (Name and City):\n")
print(df[["Name", "City"]])
```

```python
df_no_city = df.drop("City", axis=1)

print("DataFrame After Dropping City Column:\n")
print(df_no_city)
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())

print("\nDataFrame After Filling Missing Salary:\n")
print(df)
```

```python
df = df.drop_duplicates().reset_index(drop=True)

print("\nDataFrame After Removing Duplicates:\n")
print(df)
```
